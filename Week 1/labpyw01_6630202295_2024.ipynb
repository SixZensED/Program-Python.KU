```python
#9 Grade

input_score = int(input("Score: "))

if input_score >= 90 and input_score <= 100:
    print("A")
elif input_score >= 80 and input_score <= 89:
    print("B")
elif input_score >= 70 and input_score <= 79:
    print("C")
elif input_score >= 60 and input_score <= 69:
    print("D")
else:
    print("E")
```

```python
#11
def checkleap(QYear):
  if ((QYear % 400 == 0) or (QYear % 100 != 0) and (QYear % 4 == 0)):
    print("Leap Year")
  else:
    print("Not Leap Year")

QYear = int(input("Year: "))
checkleap(QYear)
```

```python
#12

def grade_cal(args):
  if args >= 90 and args <= 100:
    print("A")
  elif args >= 80 and args <= 89:
    print("B")
  elif args >= 70 and args <= 79:
    print("C")
  elif args >= 60 and args <= 69:
    print("D")
  else:
    print("E")

grade = int(input("Your Grade : "))
grade_cal(grade)
```
